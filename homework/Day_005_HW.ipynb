```python
# 我要找 CODE_GENDER = F 且 NAME_CONTRACT_TYPE = Revolving loans

condition = np.logical_and(app_train['CODE_GENDER'] == 'F',app_train['NAME_CONTRACT_TYPE'] == 'Revolving loans')
```

```python
train_data[condition]
```

```python
# 算出這群人 AMT_INCOME_TOTAL 的平均
mean = train_data[condition]['AMT_INCOME_TOTAL'].mean()
print('平均 :',mean)

# 算出這群人 AMT_INCOME_TOTAL 的標準差
std = train_data[condition]['AMT_INCOME_TOTAL'].std()
print('標準差 :',std)
```

```python
# 計算男生和女生各多少人
male_count = (app_train['CODE_GENDER'] == 'M').sum()
print(male_count)
female_count = (app_train['CODE_GENDER'] == 'F').sum()
print(female_count)
N_count = (app_train['CODE_GENDER'] == 'XNA').sum()
print(N_count)
```

```python
# 畫出圖
df = pd.DataFrame([male_count,
                   female_count,
                   N_count],
                   index = ['M', 'F', 'XNA'],
                   columns = pd.Index(['Sum']))
df.plot(kind='bar')
```
